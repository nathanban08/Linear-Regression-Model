```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

x = np.random.rand(100,1)
y = 4+3*x+ np.random.rand(100,1)

plt.scatter(x,y)
plt.title("Data")
#plt.plot(x,y, color='red')
plt.show()
```

```python
m = np.random.randn(1)
b = np.random.randn(1)

learning_rate = 0.1
epochs = 1001
```

```python
n = len(x)

for epoch in range(epochs):

    y_pred = m*x + b

    #Loss (Mean Squared Error)
    loss = (1/n)*np.sum((y_pred - y)**2)

    #setting up variables for minimizing loss
    dm = (2/n)*np.sum((y_pred - y)*x)
    db = (2/n)*np.sum(y_pred - y)

    #Minimizing Loss
    m -= (learning_rate * dm)
    b -= (learning_rate * db)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")
```
